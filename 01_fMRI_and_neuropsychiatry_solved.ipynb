We now need to apply some basic filtering. The TR is the repetition time, i.e., the interval between each time point. Can you find the correct TR for the dataset here in the article: https://www.nature.com/articles/s41597-022-01231-7

```python
timeseries = timeseries_df.to_numpy().T # we need to transpose for the filter (N time points x N regions)

TR = 2530 #insert the correct number here (from the article)

# filtering between 0.01 and 0.15 Hz
n_order = 3
low_f = 0.01
high_f = 0.2
b,a = spsg.iirfilter(n_order,[low_f,high_f],btype='bandpass',ftype='butter')
timeseries_filt = spsg.filtfilt(b,a,timeseries, axis=-1)

timex = np.arange(0, timeseries_df.shape[0]*TR,TR)/1000
```

This is the time series of the first brain region before filtering.

```python
plt.figure(figsize = (10,5))
plt.plot(timex, timeseries[0,:])
plt.plot(timex, timeseries_filt[0,:])
```

### Calculating Connectivity


We now have time series from 156regions, so we can see which regions have similar activity over time, i.e., are functionally connected to each other. This does not mean that they are directly connected via fibers, but they can be also connected via another region.

In our seminar today, we will focus on the ROI-based connectivity, i.e. we will see how each region (of interest = ROI) from a parcellation is functionally connected with each other. We will calculate a *full connectivity matrix* by computing the correlation between *all pairs of ROIs* in our parcellation scheme!

The simplest way to compute connectivity is to calculate the Pearson correlation. Find the function for the Pearson correlation in the numpy library.

Like the masker, we need to make an object that will calculate connectivity for us.

```python
fc = np.corrcoef(timeseries_filt, )
fc.shape # the shape should be N regions x N regions
```

```python
plt.figure()
plt.imshow(fc)
plt.colorbar()
plt.xlabel('ROI')
plt.ylabel('ROI')
plt.title('Functional connectivity')
```
